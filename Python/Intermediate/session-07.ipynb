# 56. Seaborn 
Seaborn is a powerful Python data visualization library built on top of Matplotlib. It provides a high-level interface for creating attractive and informative statistical graphics. 

***

# 57. Introduction to Seaborn
- Built on Matplotlib but with simpler syntax
- Beautiful default styles and color palettes
- Excellent support for pandas DataFrames
- Specialized functions for statistical visualizations
- Easy to create complex visualizations with minimal code

***

# 58. Main Plot Types

***

## 58-1. Relational Plots
- scatterplot(): Relationship between two numerical variables
- lineplot(): Relationship with continuity (typically over time)

```python
import seaborn as sns
tips = sns.load_dataset("tips")
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="time")
```

***

## 58-2. Distribution Plots
- histplot(): Histogram of a single variable
- kdeplot(): Kernel Density Estimate plot
- ecdfplot(): Empirical Cumulative Distribution Function
- rugplot(): Marginal ticks for a distribution

```python
sns.histplot(data=tips, x="total_bill", kde=True)
```

***

## 58-3. Categorical Plots
- barplot(): Point estimates and confidence intervals
- countplot(): Count of observations in each category
- boxplot(): Five-number summary of distribution
- violinplot(): Combination of boxplot and KDE
- stripplot(): Scatterplot for categorical data
- swarmplot(): Non-overlapping points for categorical data

```python
sns.boxplot(data=tips, x="day", y="total_bill", hue="smoker")
```

***

## 58-4. Regression Plots
- regplot(): Linear regression model fit
- lmplot(): Combines regplot() with FacetGrid

```python
sns.regplot(data=tips, x="total_bill", y="tip")
```

***

## 58-5. Matrix Plots
- heatmap(): Rectangular data as color-encoded matrix
- clustermap(): Hierarchically-clustered heatmap

```python
flights = sns.load_dataset("flights")
flights_matrix = flights.pivot("month", "year", "passengers")
sns.heatmap(flights_matrix, annot=True, fmt="d")
```

***

## 58-6. Multi-plot Grids
- FacetGrid: Subplot grid for plotting conditional relationships
- PairGrid: Grid of subplots for pairwise relationships
- jointplot(): Joint distribution plot with marginal distributions

```python
g = sns.FacetGrid(tips, col="time", row="smoker")
g.map(sns.scatterplot, "total_bill", "tip")
```

# 62. When to Use Seaborn vs. Matplotlib
Use Seaborn when:
- You need quick, beautiful statistical visualizations
- Working with pandas DataFrames
- Need complex visualizations with minimal code
- Want consistent styling across plots

Use Matplotlib when:
- You need maximum customization
- Creating non-standard plot types
- Need fine-grained control over plot elements

***

# 63. Example

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Loading dataset
diamonds = sns.load_dataset("diamonds")

# Setting style
sns.set_theme(style="whitegrid", palette="muted")

# Creating plot
g = sns.displot(
    data=diamonds,
    x="price", col="cut", row="color",
    kind="kde", height=3, facet_kws=dict(margin_titles=True),
)

# Customizing
g.set_axis_labels("Price (USD)", "Density")
g.set_titles(col_template="{col_name} cut", row_template="{row_name} color")
plt.suptitle("Diamond Price Distribution by Cut and Color", y=1.02)

# Saving
plt.savefig("diamonds.png", dpi=300, bbox_inches="tight")
plt.show()
```
