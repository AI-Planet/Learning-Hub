# 47. Matplotlib 
Matplotlib is the most widely used Python library for data visualization and plotting. It provides an object-oriented API for embedding plots into applications and creating publication-quality figures in a variety of formats.

***

# 48. Basic Plotting

***

## 48-1. Simple Line Plot

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.title("Sine Wave")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.grid(True)
plt.show()  
```

***

## 48-2. Scatter Plot

```python
x = np.random.rand(50)
y = np.random.rand(50)
colors = np.random.rand(50)
sizes = 1000 * np.random.rand(50)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5)
plt.colorbar()  # Adds color bar
plt.show()
```

***

## 48-3. Bar Chart

```python
labels = ['A', 'B', 'C', 'D']
values = [10, 30, 20, 25]

plt.bar(labels, values)
plt.show()
```

***

# 49. Figure and Axes Objects
The object-oriented approach is preferred for more control:

```python
fig, ax = plt.subplots()  
ax.plot(x, y, label='sin(x)')  
ax.set_title("Sine Wave")  
ax.set_xlabel("x")  
ax.set_ylabel("sin(x)")  
ax.legend()  
plt.show()
```

***

## 49-1. Figure Customization

```python
fig = plt.figure(figsize=(8, 6), dpi=100, facecolor='lightgray')
ax = fig.add_subplot(111)  # 1 row, 1 column, 1st subplot
ax.plot(x, y)
fig.suptitle("Main Title", fontsize=16)
plt.show()
```

# 50. Common Plot Types

***

## 50-1. Histogram

```python
data = np.random.randn(1000)
plt.hist(data, bins=30, density=True, alpha=0.7, color='blue')
plt.show()
```

## 50-2. Pie Chart

```python
sizes = [15, 30, 45, 10]
labels = ['A', 'B', 'C', 'D']
explode = (0, 0.1, 0, 0)  

plt.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
plt.axis('equal')  
plt.show()
```

## 51-1. Line Styles and Markers

```python
x = np.linspace(0, 10, 20)
plt.plot(x, x, '-', label='solid')  # solid
plt.plot(x, x+1, '--', label='dashed')  # dashed
plt.plot(x, x+2, ':', label='dotted')  # dotted
plt.plot(x, x+3, '-.', label='dashdot')  # dashdot
plt.plot(x, x+4, 'o', label='circles')  # circles
plt.legend()
plt.show()
```

# 52. Subplots and Multiple Axes

***

## 52-1. Basic Subplots

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))
ax1.plot(x, np.sin(x))
ax2.plot(x, np.cos(x))
plt.show()
```

***

## 52-2. GridSpec for Complex Layouts

```python
fig = plt.figure(figsize=(10, 8))
gs = fig.add_gridspec(2, 2)

ax1 = fig.add_subplot(gs[0, 0])
ax2 = fig.add_subplot(gs[0, 1])
ax3 = fig.add_subplot(gs[1, :])

ax1.plot(x, np.sin(x))
ax2.plot(x, np.cos(x))
ax3.plot(x, np.tan(x))
plt.tight_layout()
plt.show()
```

***

## 52-3. Insets

```python
fig, ax = plt.subplots()
ax.plot(x, np.sin(x))

# Create inset
ax_inset = ax.inset_axes([0.6, 0.6, 0.3, 0.3])
ax_inset.plot(x, np.cos(x))
plt.show()
```

# 53. 3D Plotting

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```

# 54. Animations

```python
from matplotlib.animation import FuncAnimation

fig, ax = plt.subplots()
x = np.linspace(0, 2*np.pi, 100)
line, = ax.plot(x, np.sin(x))

def update(frame):
    line.set_ydata(np.sin(x + frame/10))
    return line,

ani = FuncAnimation(fig, update, frames=100, interval=50)
plt.show()   
```
